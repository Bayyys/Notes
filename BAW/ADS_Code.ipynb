```python
side_length1 = math.sqrt(target_area_1) # 计算第一个谐振器边长
points_1=[] # 用于存储第一个谐振器的顶点坐标
x1,y1=0,0   # 第一个顶点坐标
points_1.append((x1,y1))    # 将第一个顶点坐标存入列表
x2,y2=0,side_length1    # 第二个顶点坐标
points_1.append((x2,y2))    # 将第二个顶点坐标存入列表
x3,y3=side_length1,side_length1 
points_1.append((x3,y3))
x4,y4=side_length1,0
points_1.append((x4,y4))
angle_range1 = side_length1/math.tan(85/360*2*math.pi)  # 计算第一个谐振器的角度范围

x1 = random.uniform(0.8*angle_range1,0+angle_range1)    # 随机生成第一个谐振器的第一个顶点坐标
y1 = 0
a = math.dist((x1,0),(0,0))
x4 = random.uniform(x4-angle_range1,x4-0.2*angle_range1)
y4 = 0
a = a + math.dist((x4,0),(side_length1,0))
y5 = side_length1 + a
x5 = random.uniform(0.3*side_length1,0.7*side_length1)
points_1[0]=x1,y1
points_1[3]=x4,y4
points_1.insert(2,(x5,y5))


slope = side_length1 / (points_1[3][0]-points_1[4][0])
def calculate_polygon_area(vertices):
    n = len(vertices)
    if n<3:
        return 0.0
    area = 0.0
    for i in range(n):
        x1,y1 = vertices[i]
        x2,y2 = vertices[(i + 1)%n]
        print(vertices[i])
        area += (x1*y2 -x2 *y1)
    area = abs(area)/2.0
    return area
```
